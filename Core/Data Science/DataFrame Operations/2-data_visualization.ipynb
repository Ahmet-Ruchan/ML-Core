```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
random_x = np.random.randn(1000)
random_y = np.random.randn(1000)

#how can i different colors based on density
plt.figure(figsize=(10, 6)) # Set figure size for better visibility
plt.scatter(random_x, random_y, alpha=0.7, color='blue', edgecolors='yellow', s=40)
plt.title('Scatter Plot of Random Points')
plt.xlabel('Random X')
plt.ylabel('Random Y')
plt.show()
```

```python
# plt.subplot()
plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st subplot

plt.hist(random_x, bins=20, color='blue', alpha=0.7)
plt.title('Histogram of Random X')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd subplot

plt.hist(random_y, bins=70, color='green', alpha=0.7) # Change color to green for distinction bins=30 for more bins
plt.title('Histogram of Random Y')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout() # Adjust layout to prevent overlap
plt.show()
```

```python
df_merge1 = pd.read_csv('7-merge_data1.csv')
df_merge2 = pd.read_csv('7-merge_data2.csv')
```

```python
df_merge1
```

```python
df_merge2
```

```python
# merge - outer join

df_merged_outer = pd.merge(df_merge1, df_merge2, on='Employee_ID', how='outer')
df_merged_outer
```

```python
df_merged_left = pd.merge(df_merge1, df_merge2, on='Employee_ID', how='left')
df_merged_left
```

```python
df_merged_right = pd.merge(df_merge1, df_merge2, on='Employee_ID', how='right')
df_merged_right
```

```python
df_merged_inner = pd.merge(df_merge1, df_merge2, on='Employee_ID', how='inner')
df_merged_inner
```

# Apply function to a column

```python
def salary_category(salary):
    if salary < 50000:
        return 'Low'
    elif 50000 <= salary < 80000:
        return 'Medium'
    else:
        return 'High'
```

```python
df['Salary_Category'] = df['Salary'].apply(salary_category)
df
```

```python
df['Name'] = df['Name'].apply(lambda name: name.replace('_', ''))
df
```
